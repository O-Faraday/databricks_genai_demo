
# Unity Catalog configuration

This notebook is used to centralize Unity Catalog parameters.

## Purpose
- Defines the catalog and schema to be used across all notebooks in the project
- Ensures consistent namespace usage throughout the demo
- Acts as a single source of truth for Unity Catalog configuration

## Workspace Configuration
It's used to work in the dedicated catalog.schema namespace.

This approach allows for:
- Easy environment switching (dev, test, prod)
- Centralized configuration management
- Simplified namespace changes across multiple notebooks

```python
# Define the Unity Catalog namespace
# Catalog: Top-level container for organizing data assets
catalog = "demo"

# Schema: Logical grouping of tables, views, and volumes within a catalog
schema = "demo"

# Set the current catalog for this Spark session
# All subsequent operations will use this catalog by default
spark.sql(f"USE CATALOG {catalog}")

# Set the current schema within the selected catalog
# All tables and volumes will be created in this schema
spark.sql(f"USE SCHEMA {schema}")

# Note: These variables can be referenced in other notebooks using %run
# This ensures consistent naming across the entire project
```
